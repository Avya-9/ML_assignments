```python
import numpy as np
import pandas as pd
from itertools import combinations
```

```python
df=pd.read_csv('/Users/avyaahuja/Machine Learning/.venv/Grocery Products Purchase.csv')
df.head()
```

```python
def c_support(item,transactions):
    count = sum([1 for t in transactions if set(item).issubset(set(t))])
    return count / len(transactions)
```

```python
def c_confidence(support_dict,X,itemset):
    support_itemset = support_dict[itemset]
    support_X = support_dict[X]
    confidence = support_itemset / support_X if support_X > 0 else 0
    return confidence
```

```python
def association(transactions, min_support=0.2, min_confidence=0.5):
    unique_items = set(item for t in transactions for item in t if item is not None)
    
    itemsets = []
    for size in range(1, len(unique_items) + 1): 
        itemsets.extend(list(combinations(unique_items, size)))
    
    support_dict = {}
    for item in itemsets:
        support = c_support(item, transactions)
        if support >= min_support:
            support_dict[frozenset(item)] = support
    

    rules = []
    for itemset in support_dict.keys():
        if len(itemset) > 1:  
            for Y in itemset:
                X = itemset - frozenset([Y])
                if len(X) > 0:
                    confidence=c_confidence(support_dict,X,itemset)
                    
                    if confidence >= min_confidence:
                        rules.append((f"{set(X)} -> {Y}", support_dict[itemset], confidence))


    
    return rules
```

```python
transactions = df.apply(lambda row: [col for col in df.columns if row[col] == 1], axis=1).tolist()
print(transactions)
```

```python
thr_supp=0.3
thr_conf=0.3
rules = association(transactions, min_support=thr_supp, min_confidence=thr_conf)
for rule in rules:
    print(f"Rule: {rule[0]}, Support: {rule[1]:.2f}, Confidence: {rule[2]:.2f}")
```
