```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/Users/avyaahuja/Machine Learning/.venv/AWCustomers.csv')
df.head()
```

```python
dfsale=pd.read_csv('/Users/avyaahuja/Machine Learning/.venv/AWSales.csv')
dfsale.head()
```

```python
data=pd.merge(df,dfsale,on='CustomerID',how='inner')
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(['Title','FirstName','MiddleName','LastName','Suffix'],axis='columns',inplace=True)
data.head()
```

```python
data.drop(['AddressLine1','AddressLine2'],axis='columns',inplace=True)
```

```python
data.head()
```

```python
data.drop(['PhoneNumber'],axis='columns',inplace=True)
```

```python
data.isnull().sum()
```

```python
data['BirthDate'] = pd.to_datetime(data['BirthDate'])
```

```python
from datetime import datetime
current_date = datetime.now()
data['Age'] = current_date.year - data['BirthDate'].dt.year
data['Age'] -= ((current_date.month < data['BirthDate'].dt.month) | ((current_date.month == data['BirthDate'].dt.month) & (current_date.day < data['BirthDate'].dt.day)))

data.head()
```

```python
data['LastUpdated'] = pd.to_datetime(data['LastUpdated'])
```

```python
data['LastUpdate'] = current_date.year - data['LastUpdated'].dt.year
data['LastUpdate'] -= ((current_date.month < data['LastUpdated'].dt.month) | ((current_date.month == data['LastUpdated'].dt.month) & (current_date.day < data['LastUpdated'].dt.day)))
data.head()
```

```python
data.drop(['LastUpdated','BirthDate'],axis='columns',inplace=True)
```

```python
data.head()
```

```python
data['Male']=pd.get_dummies(data['Gender'],drop_first=True)
data['Male']=data['Male'].astype(int)
```

```python
data.head()
```

```python
Married=pd.get_dummies(data['MaritalStatus'])
Married
Married.drop(['S'],axis='columns',inplace=True)
data['Married']=Married
```

```python
data['Married']=data['Married'].astype(int)
data.head()
```

```python
data.drop(['Gender','MaritalStatus'],axis='columns',inplace=True)
```
